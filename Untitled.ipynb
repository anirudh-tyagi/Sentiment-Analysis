```python
from wordcloud import WordCloud
```

```python
reviews = " ".join([review for review in data['verified_reviews']])

wc = WordCloud(background_color='white', max_words=100)
plt.figure(figsize=(20,20))
plt.imshow(wc.generate(reviews))
plt.title('WordCloud for all reviews', fontsize=8)
plt.axis('off')
plt.show()
```

```python
neg_reviews = " ".join([reviews for review in data[data['feedback']==0]['verified_reviews']])
neg_reviews= neg_reviews.lower().split()
```

```python
pos_reviews = " ".join([reviews for review in data[data['feedback']==1]['verified_reviews']])
pos_reviews= pos_reviews.lower().split()
```

```python
unique_negative = [x for x in neg_reviews if x not in pos_reviews]
unique_negative = " ".join(unique_negative)

unique_positive = [x for x in pos_reviews if x not in neg_reviews]
unique_positive = " ".join(unique_positive)
```

```python
rew=[neg_reviews, pos_reviews, unique_negative, unique_positive]
titles = ['Wordcloud for negative reviews', 'Wordcloud for positive reviews', 
          'Wordcloud for unique positive words', 'Wordcloud for unique negative words']
```

```python
wc = WordCloud(background_color='white', max_words=50)
```

```python
plt.figure(figsize=(12, 12))

for i, text in enumerate(data):
    # Generate the word cloud for each text
    wordcloud = wc.generate(text)
    
    # Plot the word cloud
    plt.subplot(2, 2, i+1)  # Create a 2x2 grid
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.title(titles[i], fontsize=14)  # Set dynamic title
    plt.axis('off')  # Hide the axis

# Adjust layout to prevent overlap
plt.tight_layout()
plt.show()
```
